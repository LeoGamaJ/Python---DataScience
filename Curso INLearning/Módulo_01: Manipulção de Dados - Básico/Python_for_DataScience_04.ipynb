```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

```python
dados = {
    "Coluna 1": [1, 1, 2, 2, 3, 3, 3],
    "Coluna 2": ["a", "a", "b", "b", "c", "c", "c"],
    "Coluna 3": ["A", "A", "B", "B", "C", "C", "C"]
}

df = DataFrame(dados)
df
```

```python
# Linhas com valores repetidos

df.duplicated()
```

```python
# Eliminando valores repetidos

df.drop_duplicates()
```

```python
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

```python
# df1

df1 = pd.DataFrame(np.arange(36).reshape(6,6))
df1
```

```python
# df2

df2 = pd.DataFrame(np.arange(15).reshape(5,3))
df2
```

```python
# Concatenando

pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], axis=1)
```

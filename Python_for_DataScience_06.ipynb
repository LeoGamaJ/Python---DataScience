# Transformando dados

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.arange(36).reshape(6,6))
df
```

```python
# Remover Dados
# Linhas

df.drop([0, 2])
```

```python
# Remover Dados
# Colunas

df.drop([1, 2], axis=1)
```

```python
serie = pd.Series(np.arange(6))
serie.name = "nova_variavel"

serie
```

```python
# unindo DataFrame com join 

novo_df = pd.DataFrame.join(df, serie)
novo_df
```

```python
# Ordenando Objetos

df_ordenado = novo_df.sort_values(by=["nova_variavel"], ascending=False)
df_ordenado
```

```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

```python
# Criando serie para exemplo

valores_nan = np.nan

serie = Series(["linha 1", "linhda2", valores_nan, "linhda 4", "linhda 5", "linha 6", valores_nan, "linha 8" ])
serie
```

```python
# Exibindo valores ausentes NaN
serie.isnull()
```

```python
# Criando df com valores NaN

np.random.seed(25)

df = DataFrame(np.random.rand(36).reshape((6,6)))
df.loc[3:5, 0] = valores_nan
df.loc[1:4, 0] = valores_nan

df
```

```python
# Preenchendo valores NaN com 0

df_preenchido = df.fillna(0)
df_preenchido
```

```python
dicio = {0: 0.1, 5: 1.25}
df_preenchido = df.fillna(dicio)
df_preenchido
```

```python
# Contando os valores NaN

df.isnull().sum()
```

```python
# Conjunto de linhas e colunas que não apresentam valores em branco

df.dropna()
```
